#        EDA (Exploratory Data Analysis)

```python
# Map Yes/No-like binaries to 1/0
binary_map = {
    'Yes': 1, 'No': 0, 'No internet service': 0,
    'Male': 1, 'Female': 0,
    'True': 1, 'False': 0
}

def map_binary(col):
    unique_vals = set(df[col].dropna().unique())
    # Identify classic binaries
    candidates_yesno = {'Yes', 'No', 'No internet service'}
    candidates_gender = {'Male', 'Female'}
    candidates_tf = {'True', 'False'}
    
    if (unique_vals.issubset(candidates_yesno) or 
        unique_vals.issubset(candidates_gender) or 
        unique_vals.issubset(candidates_tf)):
        return df[col].map(binary_map).astype('int')
    
    return df[col]

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = map_binary(col)

df.sample(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def hist_boxplot(cols):
    # Create subplots with 2 rows (hist + boxplot) and len(cols) columns
    fig, axes = plt.subplots(2, len(cols), figsize=(5 * len(cols), 8))

    for i, col in enumerate(cols):
        # Histogram
        sns.histplot(
            data=df, x=col, ax=axes[0, i], kde=True, color="#4C7280"
        )
        axes[0, i].set_title(f'Histogram of {col}', color="#4C7280")

        # Boxplot
        sns.boxplot(
            data=df, x=col, ax=axes[1, i], color="#4C7280"
        )
        axes[1, i].set_title(f'Boxplot of {col}', color="#4C7280")

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_piecharts(df, cols):
    n = len(cols)
    fig, axes = plt.subplots(1, n, figsize=(6*n, 6))

    if n == 1:  # If only one column, axes is not iterable
        axes = [axes]

    for i, col in enumerate(cols):
        counts = df[col].value_counts()
        percentages = counts / counts.sum() * 100

        axes[i].pie(
            counts,
            labels=[f"{idx} ({val:.1f}%)" for idx, val in zip(counts.index, percentages)],
            autopct='%1.1f%%',
            startangle=90,
            colors=plt.cm.Paired.colors
        )
        axes[i].set_title(f"Pie Chart of {col}", fontsize=14)

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Exclude specific columns
exclude_cols = ['MonthlyCharges', 'tenure', 'customerID', 'TotalCharges']
columns = [col for col in df.columns if col not in exclude_cols]

# Set up figure: 3 charts per row
n_cols = 3
n_rows = int(np.ceil(len(columns) / n_cols))
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows*5))
axes = axes.flatten()

# Loop through columns and create lollipop charts
for i, col in enumerate(columns):
    counts = df[col].value_counts()
    
    # Create lollipop chart
    markerline, stemlines, baseline = axes[i].stem(counts.index, counts.values, basefmt=" ")
    
    # Increase marker size
    markerline.set_markersize(12)
    markerline.set_markerfacecolor('red')
    
    # Show count above each lollipop
    for x, y in zip(counts.index, counts.values):
        axes[i].text(x, y + 0.5, str(y), ha='center', va='bottom', fontsize=10)
    
    axes[i].set_title(f'{col} Count')
    axes[i].set_ylabel('Count')
    axes[i].set_xlabel(col)

# Remove empty subplots if any
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
